**基于物品的协同过滤(ItemCF)** 作为核心模型，这是最简单且有效的推荐算法



核心思想：根据物品之间的相似度来推荐。如果用户A喜欢物品1，而物品1与物品2相似，那么用户A可能也会喜欢物品2。

一、建立交互矩阵
所以我们编写程序的首要目的是，通过分析从数据库获取的用户文档交互数据，建立用户喜爱的文档矩阵。

二、建立物品相似性矩阵
根据生成的交互矩阵，对数据库中的所有文档做相似性分析。

三、基于相似性对未浏览的文档进行评分
得到一系列的评分数据，根据用户未浏览的文档的评分数据得到用户可能喜欢的文档的Top10

***TtemCF方法如何分析物品的相似性***

**基本思想**

一、余弦相似性

将物品表示为用户评分向量，计算两个向量的夹角余弦值
通过分析两个向量在高维空间上的线性相似程度来判断两个事物是否相似。

*（我们也可以通过对评分向量做去中心化处理，得到调整余弦相似性，用于减少极高分和极低分对结果的影响）*

```python
"""
sim(i, j) = |U(i) ∩ U(j)| / |U(i) ∪ U(j)|
U(i) 表示对物品i有过行为的用户集合
物品A被用户 {1, 2, 3, 5} 浏览过
物品B被用户 {2, 3, 4, 5} 浏览过
"""
```

一、Jaccard相似性

  基于用户集合的重叠程度
  交集 = {2, 3, 5} = 3个用户
  并集 = {1, 2, 3, 4, 5} = 5个用户

  相似度 = 3/5 = 0.6

*（实际应用中常加入惩罚因子，避免共同用户太少导致的偶然高相似度：
  $sim(i, j) = (|U(i) ∩ U(j)| / (|U(i) ∪  U(j)| + λ)) × cosine(i, j)$
  其中λ是平滑参数（通常取50-100）*

```python
"""
预测用户对未浏览文档的兴趣评分

        预测公式:
        r̂(u,i) = Σ[sim(i,j) × r(u,j)] / Σ|sim(i,j)|
"""
```

三、得到兴趣评分

做加权求和并除上平均相似度之和，得到兴趣分数
